```python
import pandas as pd
import re
from rapidfuzz import fuzz, process

# --- STEP 1: Keyword-based base title extraction ---
def strip_after_keyword(title):
    t = title.strip()
    base = re.split(
        r"\b(season|after story|ova|movie|special|part|episode|ep|final|arc|sp|2nd|second season)\b",
        t,
        flags=re.IGNORECASE
    )[0].strip()
    base = re.sub(r"[\s:!?.-]+$", "", base)  # clean trailing punctuation
    return base
```

```python
# --- STEP 2: Normalize strings for fuzzy matching ---
def normalize_title(title):
    title = title.lower()
    title = re.sub(r"[^\w\s]", "", title)  # remove punctuation
    title = re.sub(r"\s+", " ", title)     # collapse spaces
    return title.strip()
```

```python
# --- STEP 3: Hybrid grouping logic ---
def group_titles(df, threshold=90):
    """
    Groups similar titles using fuzzy matching.
    threshold: similarity score (0-100) above which titles are considered the same.
    """
    grouped_ids = set()
    base_to_group = {}
    normalized_titles = df["base_title_norm"].tolist()

    for idx, base_norm in enumerate(normalized_titles):
        if idx in grouped_ids:
            continue

        # Find close matches
        matches = process.extract(
            base_norm,
            normalized_titles,
            scorer=fuzz.ratio,
            score_cutoff=threshold
        )

        # Get all indexes of matches
        match_indexes = [i for i, _ in [(m[2], m[1]) for m in matches]]
        match_indexes.append(idx)

        # Assign group name (first title in group)
        group_name = df.iloc[idx]["base_title_norm"]
        for mi in match_indexes:
            base_to_group[mi] = group_name
            grouped_ids.add(mi)

    df["title_group"] = df.index.map(base_to_group)
    return df
```

```python
df["base_title"] = df["title"].apply(strip_after_keyword)
df
```

```python
df["base_title_norm"] = df["base_title"].apply(normalize_title)
df
```

```python
df[df['title_group'] == 'haikyuu']
```

```python
df = group_titles(df, threshold=90)
df
```

```python
df = df.sort_values("mal_id").drop_duplicates(subset=["title_group"], keep="first")
```

```python
df = df.sort_values(by = 'score', ascending = False)
```

```python
df = df.reset_index(drop = True)
```
